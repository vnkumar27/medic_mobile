```python
ses_vars = ['water_source', 'toilets', 'cooking_fuel', 'floor_material', 'food_security']
ses = df[['patient_id'] + ses_vars]
ses.shape
ses.sample(10)
ses.info()
```

```python
ses_cleaned = ses[ses[ses_vars].any(axis=1)]
ses_cleaned
```

```python
for ses in ses_vars:
    ses_cleaned[ses] = ses_cleaned[ses].replace(99, np.nan) # 99 is equivalent to no response, so nulling it.
    print(ses_cleaned[ses].value_counts())
    print()
ses_cleaned = ses_cleaned[ses_cleaned[ses_vars].any(axis=1)] # Remaking df after removing extra null values
```

```python
# Cleaning ses columns that weren't finished mapping
ses_cleaned['toilets'] = ses_cleaned['toilets'].replace({61: "no toilet", 
                                                         51: "toilet"})
ses_cleaned['toilets'].value_counts()
```

```python
ses_cleaned['cooking_fuel'] = ses_cleaned['cooking_fuel'].replace({8: "wood/straw",
                                                                   9: "herbs",
                                                                   7: "charcoal",
                                                                  11: "bouse"})
ses_cleaned['cooking_fuel'].value_counts()
```

```python
ses_cleaned['floor_material'] = ses_cleaned['floor_material'].replace({11: "earth/sand",
                                                                       12: "bouse",
                                                                       22: "fins/bamboo",
                                                                       34: "cement"})
ses_cleaned['floor_material'].value_counts()
```

```python
ses_cleaned['food_security'] = ses_cleaned['food_security'].replace({2: "no",
                                                                     1: "yes"})
ses_cleaned['food_security'].value_counts()
```
